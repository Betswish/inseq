```python
import inseq
import logging

logger = logging.getLogger(__name__)
logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s - %(name)s - %(message)s",
    datefmt="%m/%d/%Y %H:%M:%S",
)

model = inseq.load("Helsinki-NLP/opus-mt-en-it", "integrated_gradients")
txt = ["Very short", "Hello world, today is a good day!", "How can I help you?"]
out = model.attribute(txt, references=None, attr_pos_end=None, return_convergence_delta=True, n_steps=5)
out[1].heatmap()
```
